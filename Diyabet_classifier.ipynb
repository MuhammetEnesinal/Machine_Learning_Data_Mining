```python
import numpy as np
from sklearn.model_selection import train_test_split

df = df[df['BloodPressure'] != 0]


X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Train-Test Split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)


columns_to_fix = ["Glucose", "BMI","SkinThickness","Insulin"]

for col in columns_to_fix:
    # SADECE X_train içindeki 0 olmayan değerlerin medyanını buluyoruz
    train_median = X_train[X_train[col] != 0][col].median()
    
    # Bu medyan değerini kullanarak hem Train hem de Test setini dolduruyoruz
    X_train[col] = X_train[col].replace(0, train_median)
    X_test[col] = X_test[col].replace(0, train_median)

```


LOGİSTİC REGRESYON

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

logistic_regresyon =LogisticRegression()

logistic_regresyon_param = {
    "penalty": ["l1", "l2", "elasticnet", None], 
    "C": [0.1, 1, 10, 100],                       
    "solver": ["liblinear", "saga"]              
}

grid=GridSearchCV(estimator=logistic_regresyon,cv=3,param_grid=logistic_regresyon_param,n_jobs=-1,verbose=1)


grid.fit(X_train,y_train)


# 1. En iyi parametreleri gör (C, penalty ve solver hangisi seçildi?)
print("En iyi parametreler:", grid.best_params_)

# 2. En iyi skoru gör (Başarı oranı kaç?)
print("En iyi skor:", grid.best_score_)

# 3. En iyi modeli doğrudan al ve kullan
en_iyi_model = grid.best_estimator_
```

```python
logistic_regresyon =LogisticRegression(C=1,penalty="l1",solver="liblinear")

logistic_regresyon.fit(X_train,y_train)
y_pred_logistic=logistic_regresyon.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_logistic))
print(accuracy_score(y_test, y_pred_logistic))
print(confusion_matrix(y_test, y_pred_logistic))
```

Destek Vektör Makineleri

```python
from sklearn import svm
svc_classifier=svm.LinearSVC()

svc_classifier_param = {
    "penalty": ["l2"],            # LinearSVC'de çoğu durumda l2 kullanılır
    "loss": ["hinge", "squared_hinge"], 
    "C": [0.1, 1, 5, 10],         # 0 yerine 0.1 gibi küçük bir değerle başla
    "class_weight": ["balanced", None],
    "max_iter": [2000]            # Yakınsama (convergence) hatası almamak için artırmak iyidir
}

grid=GridSearchCV(estimator=svc_classifier,cv=3,param_grid=svc_classifier_param,n_jobs=-1,verbose=1)

grid.fit(X_train,y_train)

# 1. En iyi parametreleri gör (C, penalty ve solver hangisi seçildi?)
print("En iyi parametreler:", grid.best_params_)

# 2. En iyi skoru gör (Başarı oranı kaç?)
print("En iyi skor:", grid.best_score_)

# 3. En iyi modeli doğrudan al ve kullan
en_iyi_model = grid.best_estimator_
```

```python
svc_classifier=svm.LinearSVC(C=10,class_weight=None,loss="squared_hinge",max_iter=2000,penalty="l2")

svc_classifier.fit(X_train,y_train)
y_pred_svc=svc_classifier.predict(X_test)

print(classification_report(y_test, y_pred_svc))
print(accuracy_score(y_test, y_pred_svc))
print(confusion_matrix(y_test, y_pred_svc))
```

BernoulliNB

```python
from sklearn.naive_bayes import BernoulliNB
nb_bernoulli=BernoulliNB()
bernoulli_nb_param = {
    'alpha': [0.01, 0.1, 0.5, 1.0, 10.0], # Laplace düzeltmesi (Smoothing)
    'binarize': [0.0, 0.5, 1.0]           # Sayıları 0/1 yapmak için eşik değeri
}
grid_bnb = GridSearchCV(estimator=nb_bernoulli, param_grid=bernoulli_nb_param, cv=3, n_jobs=-1, verbose=1)
grid_bnb.fit(X_train,y_train)

# 1. En iyi parametreleri gör (C, penalty ve solver hangisi seçildi?)
print("En iyi parametreler:", grid_bnb.best_params_)

# 2. En iyi skoru gör (Başarı oranı kaç?)
print("En iyi skor:", grid_bnb.best_score_)

# 3. En iyi modeli doğrudan al ve kullan
en_iyi_model = grid_bnb.best_estimator_
```

```python
nb_bernoulli=BernoulliNB(alpha=0.01, binarize=1.0)

nb_bernoulli.fit(X_train,y_train)
y_pred_svc=nb_bernoulli.predict(X_test)

print(classification_report(y_test, y_pred_svc))
print(accuracy_score(y_test, y_pred_svc))
print(confusion_matrix(y_test, y_pred_svc))
```

KNeighborsClassifier 

```python
from sklearn.neighbors import KNeighborsClassifier

knn_param = {
    "n_neighbors": [3, 5, 7, 9, 11, 15], # Bakılacak komşu sayısı
    "weights": ["uniform", "distance"],  # Komşuların ağırlığı (Eşit veya Mesafeye göre)
    "metric": ["euclidean", "manhattan"] # Mesafe hesaplama yöntemi
}

knn = KNeighborsClassifier()

grid_knn = GridSearchCV(
    estimator=knn,
    param_grid=knn_param,
    cv=3, 
    n_jobs=-1, 
    verbose=1
)

grid_knn.fit(X_train, y_train)

print("En iyi KNN parametreleri:", grid_knn.best_params_)
print("En iyi KNN skoru:", grid_knn.best_score_)
```

```python
knn = KNeighborsClassifier(metric="euclidean",n_neighbors=15,weights="distance")
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)

print(classification_report(y_test, y_pred_knn))
print(accuracy_score(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))
```
